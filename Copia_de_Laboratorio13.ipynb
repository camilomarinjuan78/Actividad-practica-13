Prompt 1: Análisis de Sentimiento con TextBlob

Instrucciones:
Utiliza la librería TextBlob para analizar el sentimiento de las siguientes frases. Debes interpretar el sentimiento de las frases como positivo, negativo o neutral en base a su polaridad.

Frase 1: The new compressor design provides excellent cooling efficiency and operates very quietly.
Frase 2: The evaporator unit is leaking refrigerant and shows a terrible drop in pressure.
Frase 3: The technician recorded the internal temperature at five degrees Celsius during the test.

Objetivo:
Para cada frase, analiza la polaridad y asigna el sentimiento correspondiente:

Positivo si la polaridad es mayor que 0.
Negativo si la polaridad es menor que 0.
Neutral si la polaridad es 0.

```python
from textblob import TextBlob

frases = [
    "The new compressor design provides excellent cooling efficiency and operates very quietly.",
    "The evaporator unit is leaking refrigerant and shows a terrible drop in pressure.",
    "The technician recorded the internal temperature at five degrees Celsius during the test."
]

for i, frase in enumerate(frases, 1):
    analysis = TextBlob(frase)
    polarity = analysis.sentiment.polarity

    if polarity > 0:
        sentiment = "Positivo"
    elif polarity < 0:
        sentiment = "Negativo"
    else:
        sentiment = "Neutral"

    print(f"Frase {i}: {sentiment} (Polaridad: {polarity})")
```

Frase 1: "The new compressor design provides excellent cooling efficiency and perfect temperature stability."
Traducción: "El nuevo diseño del compresor proporciona una excelente eficiencia de enfriamiento y una perfecta estabilidad de temperatura."
Frase 2: "The refrigeration system is failing the pressure tests due to a terrible refrigerant leak."
Traducción: "El sistema de refrigeración está fallando las pruebas de presión debido a una terrible fuga de refrigerante."
Frase 3: "The technician is monitoring the evaporation pressure and recording the data in the log."
Traducción: "El técnico está monitoreando la presión de evaporación y registrando los datos en el cuaderno de bitácora."

Prompt 2: Entrenamiento de Modelo de Clasificación de Sentimientos con Naive Bayes

Instrucciones:
Entrena un modelo de clasificación de sentimientos utilizando el algoritmo Naive Bayes. Utiliza los siguientes ejemplos de frases etiquetadas y evalúa la precisión de tu modelo.

Datos de ejemplo:

Frase 1: "This app is very intuitive!"
Frase 2: "I didn't enjoy the customer service."
Frase 3: "Quite average experience."
Frase 4: "Fantastic presentation!"
Frase 5: "The food was awful."

Objetivo:

Entrena un modelo de Naive Bayes con estas frases etiquetadas.
Evalúa la precisión del modelo utilizando las frases proporcionadas.

```python
from textblob.classifiers import NaiveBayesClassifier
import nltk
nltk.download('punkt_tab') # Download the missing NLTK data

# 1. Conjunto de datos etiquetados (Datos de entrenamiento)
train_data = [
    ("This app is very intuitive!", "positivo"),
    ("I didn't enjoy the customer service.", "negativo"),
    ("Quite average experience.", "neutral"),
    ("Fantastic presentation!", "positivo"),
    ("The food was awful.", "negativo")
]

# 2. Entrenamiento del modelo
# El clasificador analiza la probabilidad de que una palabra pertenezca a cada etiqueta.
model = NaiveBayesClassifier(train_data)

# 3. Evaluación de la precisión (Accuracy)
# Medimos qué tan bien clasifica el modelo los datos originales.
accuracy = model.accuracy(train_data)

print(f"--- Resultados del Modelo ---")
print(f"Precisión obtenida: {accuracy * 100}%")

# 4. Verificación de etiquetas predichas
print(f"\n--- Clasificación detallada ---")
for frase, etiqueta_real in train_data:
    prediccion = model.classify(frase)
    print(f"Frase: '{frase}' -> Predicción: {prediccion}")
```
